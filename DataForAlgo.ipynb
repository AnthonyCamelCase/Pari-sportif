```python
match_non_fini = df[(df.Comment == 'Retired')| (df.Comment == "Walkover") | (df.Comment == "Disqualified")].index
```

```python
df = df.drop(match_non_fini)
df.index = range(len(df.index)) 
```

A partir de là c'est flou

```python
df["elo_diff"]= df["elo_winner"]-df["elo_loser"]
df["rank_diff"]= df.WRank - df.LRank
```

```python
df["Vainqueur"]= "wait"

for i in range(len(df)):
    if (df["WRank"][i] <= df["LRank"][i]):
        df["Vainqueur"][i] = "favori"
    else :
        df["Vainqueur"][i] = "challenger"
```

Quelle est vraiment la cible de notre projet ?

```python
target = df.Vainqueur
```

```python
data = df.drop("Vainqueur",axis=1)
```

Algo

```python
X_train, X_test, y_train, y_test= train_test_split(data , target, test_size = 0.3)
```

NOrmalisation, centrer et réduire :

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel="poly", gamma = 0.001)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

KNN plus proches voisins mais pas adapté à ici je pense

```python
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(
    n_neighbors = 7,
    metric = "minkowski",
    p=2)

knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

Arbre de décision

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion="entropy",max_depth=4)

dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
```

```python
pd.crosstab(y_test,y_pred,rownames=["reel"], colnames=["prédites"])
```
