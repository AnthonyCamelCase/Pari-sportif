<h2>Formule pour la corrélation entre variable qualitatives </h2> 

```python
from scipy.stats import chi2_contingency
import math

def V_Cramer(tableau, N):
    r = table.shape[1]
    k = table.shape[0]
    k_corr = k - (k-1)**2 / (N-1)
    r_corr = r - (r-1)**2 / (N-1)
    phi2 = max(0,chi2_contingency(table)[0]/N - (k-1)*(r-1) / (N-1))
    V = math.sqrt(phi2 / min(k_corr-1,r_corr-1))
    return V
```

<h2> Corrélation entre les variables qualitatives et la variable cible <b>Winner</b></h2>

```python
var_cat = ["ATP", "Location", "Tournament", "Date", "Series", "Court", "Surface", "Round", "Best of", "Wsets", "Lsets", "Comment"]
```

```python
for var in var_cat:
    print(f"corrélation entre la variable Winner et la variable {var}")
    table = pd.crosstab(df["Winner"],df[var])

    print(f"la p value est de : {chi2_contingency(table)[1]}")

    if (chi2_contingency(table)[1] <= 0.05):
        print(f"le V de Cramer est de : {V_Cramer(table,df.shape[0])}, \n")
    else:
        print("\n")
```
