```python
import pandas as pd
import numpy as np

df = pd.read_csv('data_clean2.csv',  index_col = 'Unnamed: 0')
```

```python
target = df.Vainqueur
data = df.drop('Vainqueur', axis = 1)
data.head()
```

```python
# variables Catégorielles:
cat = data.loc[:,data.dtypes==np.object]
cat = pd.get_dummies(cat)
cat.shape
```

```python
#  variables numeriques:
float_ = data.loc[:,data.dtypes== 'float64']
int_ = data.loc[:,data.dtypes== 'int64']
num = pd.concat([float_, int_], axis = 1)
num.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num)
num_scaled = pd.DataFrame(num_scaled, columns = num.columns)
num_scaled
```

```python
data = pd.concat([cat,num_scaled], axis = 1)
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state = 42, max_iter = 1000)
clf.fit(X_train, y_train)
y_pred_test = clf.predict(X_test)
y_pred_train = clf.predict(X_train)
print('score du test :', clf.score(X_test, y_test))
print('score du train :', clf.score(X_train, y_train))
matrice = pd.crosstab(y_test, y_pred_test, colnames= ['Predictions'],rownames = ['réalité'])
pd.DataFrame(matrice)
```

```python
# création du dataframe Paris

paris = df[['PS_F','B365_F','PS_C','B365_C']]
paris['proba_favoris'] = clf.predict_proba(data)[:,1]
paris['proba_chalenger'] = 1 - paris['proba_favoris']

paris
```

```python
paris['résultat'] = target
paris
```

```python
paris['meilleur_cote_fav'] = 'wait'
paris['meilleur_cote_chal'] = 'wait'
for i in range(len(paris)):
    if paris['PS_F'][i]> paris['B365_F'][i]:
        paris['meilleur_cote_fav'][i] = paris['PS_F'][i]
    else:
        paris['meilleur_cote_fav'][i] = paris['B365_F'][i]
    if paris['PS_C'][i]>paris['B365_C'][i]:
        paris['meilleur_cote_chal'][i] = paris['PS_C'][i]
    else:
        paris['meilleur_cote_chal'][i] = paris['B365_C'][i]
    
```

```python
paris.drop(columns = ['PS_F', 'B365_F', 'PS_C', 'B365_C'], inplace = True)
```

```python
paris.head()
```

```python
paris['proba_bookmaker_fav'] = 1/paris.meilleur_cote_fav
paris['proba_bookmaker_chal'] = 1/paris.meilleur_cote_chal
```

```python
paris.head()
```

```python
paris['confiance_fav'] = paris.proba_favoris/paris.proba_bookmaker_fav
paris['confiance_chal'] = paris.proba_chalenger/paris.proba_bookmaker_chal
```

```python
paris['confiance_paris'] = 'wait'
paris['choix'] = 'wait'
for i in range(len(paris)):
    if paris.confiance_fav[i]>paris.confiance_chal[i]:
        paris['confiance_paris'][i] = paris.confiance_fav[i]
        paris.loc[i, 'choix'] = 'favori'
    else:
        paris['confiance_paris'][i] = paris.confiance_chal[i]
        paris.loc[i, 'choix'] = 'challenger'
        
```

```python
paris.head()
```

```python
paris.sort_values(by = 'confiance_paris', ascending = False, inplace = True)
```

```python
paris = paris.reset_index()
```

```python
paris.drop('index', axis = 1, inplace =True)
```

```python
paris.head()
```

```python
def par(n):
    gain = 0
    for i in range(n):
        if paris['choix'][i] == paris['résultat'][i] == 'favori':
            gain = gain + 1*paris.meilleur_cote_fav[i] - 1
        elif paris['choix'][i] == paris['résultat'][i] == 'challenger':
            gain = gain + 1*paris.meilleur_cote_chal[i] - 1
        else :
            gain = gain -1
    return gain
    
```

```python
nombre_de_paris = [1,3,10,20,50,100,500,1000,5000, 10000, len(paris)]
gain_roi = []
for i in nombre_de_paris:
    gain_roi.append(par(i))
    
```
