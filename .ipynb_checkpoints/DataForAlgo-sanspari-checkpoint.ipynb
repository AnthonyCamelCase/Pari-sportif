Trouvons les meilleurs hyperparamètres :

```python
def Grid(clf, param):
    grid_clf = GridSearchCV(estimator=clf,param_grid = param)
    outer_cv = StratifiedKFold(n_splits = 3, shuffle=True)
    score = cross_val_score(grid_clf,X_train,y_train,cv = outer_cv)
    print("accuracy du modèle : ", score.mean())
    print("ecart type du modèle : ", score.std())
    grid_clf.fit(X_train,y_train)
    print("meilleur paramètre : ", grid_clf.best_params_)
    print("score train : ",round(100*grid_clf.score(X_train,y_train),2))
    print("score test : ", round(100*grid_clf.score(X_test,y_test),2))
```

Trouvez le moyen de comparer pour se faire de l'argent

```python
pari["proba_PS_Favori"]=1/pari.PS_F
```

```python
pari["proba_PS_Challenger"]=1/pari.PS_C
```

```python
pari["indice_Favori"]=pari.Proba_Favori/ pari.proba_PS_Favori
```

```python
pari["indice_Challenger"]=pari.Proba_Challenger/ pari.proba_PS_Challenger
```

```python
pari
```

```python
pari.to_csv("pari.csv")
```

```python
test = pari.drop(["Favori","Challenger","Proba_Favori","Proba_Challenger","proba_PS_Favori","proba_PS_Challenger"], axis=1)
```

```python
test= test.reset_index().drop("index",axis=1)
```

```python
test['indice'] = 'wait'
test['choix'] = 'wait'
for i in range(len(test)):
    if (test.indice_Favori[i]>test.indice_Challenger[i]):
        test['indice'][i] = test.indice_Favori[i]
        test['choix'][i] = 'favori'
    else:
        test['indice'][i] = test.indice_Challenger[i]
        test['choix'][i] = 'challenger'
    
```

```python
test.sort_values(by = 'indice', ascending = False, inplace = True)
test
```

```python
def gain(n):
    gain = 0
    argent = 0
    gain_roi = []
    for i in range(n):
        mise = 1
        if test["Vainqueur"][i] == test["choix"][i] == "favori":
                gain += mise * test.loc[i,"PS_F"]
                argent += mise * test.loc[i,"PS_F"]-1
        elif test["Vainqueur"][i] == test["choix"][i] == "challenger":
                gain += mise * test.loc[i,"PS_C"]
                argent += mise * test.loc[i,"PS_C"]-1
        else:
                argent -= mise
    
    rendement = gain / n *100
    print(f"pour {n} matchs, vous avez gagné {round(argent,2)} et le rendement est de {round(rendement,2)}% ")
    return argent
```

```python
nombre_de_paris = [1,3,10,20,50,100,500,1000,5000, 10000, len(test)]
gain_roi = []
for i in nombre_de_paris:
    gain_roi.append(gain(i))

```

```python
plt.plot(nombre_de_paris, gain_roi)
plt.xlabel('nombre de paris')
plt.ylabel('gain');
```
