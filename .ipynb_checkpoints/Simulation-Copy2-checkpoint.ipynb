```python
import pandas as pd
import numpy as np

df = pd.read_csv('data_clean2.csv',  index_col = 'Unnamed: 0')
```

```python
target = df.Vainqueur
data = df.drop('Vainqueur', axis = 1)
data.head()
```

```python
# variables Catégorielles:
cat = data.loc[:,data.dtypes==np.object]
cat = pd.get_dummies(cat)
cat.shape
```

```python
#  variables numeriques:
float_ = data.loc[:,data.dtypes== 'float64']
int_ = data.loc[:,data.dtypes== 'int64']
num = pd.concat([float_, int_], axis = 1)
num.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
num_scaled = scaler.fit_transform(num)
num_scaled = pd.DataFrame(num_scaled, columns = num.columns)
num_scaled
```

```python
data = pd.concat([cat,num_scaled], axis = 1)
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state = 42, max_iter = 1000)
clf.fit(X_train, y_train)
y_pred_test = clf.predict(X_test)
y_pred_train = clf.predict(X_train)
print('score du test :', clf.score(X_test, y_test))
print('score du train :', clf.score(X_train, y_train))
matrice = pd.crosstab(y_test, y_pred_test, colnames= ['Predictions'],rownames = ['réalité'])
pd.DataFrame(matrice)
```

```python
# création du dataframe Paris

paris = df[['PS_F','B365_F','PS_C','B365_C']]
paris['proba_favoris'] = clf.predict_proba(data)[:,1]
paris['proba_chalenger'] = 1 - paris['proba_favoris']

paris
```

```python
# Choix de décision
paris['decision'] = 'wait'
for i in range(len(paris)):
    if abs(paris.proba_favoris[i]-paris.proba_chalenger[i]) >0.2:
        paris.loc[i,'decision'] = 'oui'
    else :
        paris.loc[i,'decision'] = 'non'
paris.head(1)    
```

```python
for i in range(len(paris)):
        if paris.proba_favoris[i] > paris.proba_chalenger[i]:
            paris.loc[i,'choix'] = 'favori'
        else:
            paris.loc[i,'choix'] = 'challenger'
```

```python
paris['résultat'] = clf.predict(data);
```

```python
paris_oui = paris[paris['decision'] == 'oui']
paris_oui.reset_index(inplace = True)
paris_oui.drop('index', axis = 1, inplace=True)
paris_oui.head()
```

```python
gain = 0
for i in range(len(paris_oui)):
    
    if paris_oui.PS_F[i]>paris_oui.B365_F[i]:
        pf = paris_oui.PS_F[i]
    else:
        pf = paris_oui.B365_F[i]
        
    if paris_oui.PS_C[i]>paris_oui.B365_C[i]:
        pc = paris_oui.PS_C[i]
    else:
        pc = paris_oui.B365_C[i]
        
    if paris_oui.choix[i] == paris_oui['résultat'][i] == 'favori':
        gain = gain + 1*pf
               
    elif paris_oui.choix[i] == paris_oui['résultat'][i] == 'challenger':
        gain = gain  + 1*pc
                
    else:
        gain = gain - 1
               
```
